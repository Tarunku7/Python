```python
def convert_size(value):
    if 'M' in str(value):
        return float(value.replace('M', ''))
    elif 'k' in str(value):
        return round(float(value.replace('k', '')) / 1024, 2)  # convert KB to MB
    else:
        return np.nan  # for "Varies with device" or NaN

df["Size"] = df["Size"].apply(convert_size)
```

```python
df["Size"] = df["Size"].fillna(df["Size"].mean())
df["Size"].unique()
```

```python
df["Installs"].unique()
```

```python
char_to_remove=["+",",","$"]
cols_to_clean = ["Installs","Price"]
for i in char_to_remove:
    for j in cols_to_clean:
        df[j]=df[j].str.replace(i,"")
```

# What are the 5 most installed app in each popular Categories?

```python
dfa = df.groupby(["Category","App"])["Installs"].sum().reset_index()
dfa= dfa.sort_values("Installs", ascending = False)
apps =["GAME", "COMMUNICATION", "PRODUCTIVITY", "SOCIAL"]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))
for i, app in enumerate(apps):
    df2=dfa[dfa.Category==app]
    df3=df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data=df3, x="Installs",y="App", palette = "husl")
    plt.xlabel("Installation in Millions")
    plt.ylabel("")
    plt.title(app,size=20)
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

```python
rating =df.groupby(["Category","Installs","App"])["Rating"].sum().sort_values(ascending = False).reset_index()

top_rating_app = rating[rating.Rating == 5.0]
print("Number of 5 rated apps",top_rating_app.shape[0])
top_rating_app=top_rating_app.head(5)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(y='App', x='Installs', data=top_rating_app, palette='mako')

plt.title('Top 5 Rated Apps Compared with Installs')
plt.xlabel('Number of Installs')
plt.ylabel('App Name')
plt.xscale('log')  # Because Install numbers are usually huge
plt.tight_layout()

plt.show()
```
