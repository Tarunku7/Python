```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#rename column
df.rename(columns= {'Marital_Status':'Shaadi'})
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for i in ax.containers:
    ax.bar_label(i)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sales_gen
```

```python
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
plt.show()
```

### Age

```python
age = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for j in age.containers:
    age.bar_label(j)
```

```python
# Total Amount vs Age Group
age_amount = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
print(age_amount)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = age_amount)
plt.show()
```

```python
# Total Orders v/s Age_group

age_orders = df.groupby(['Age Group'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
print(age_orders)

sns.barplot(x = 'Age Group',y= 'Orders' ,data = age_orders)
plt.show()
```

### State

```python
# total number of orders from top 10 states

ord_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
ord_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = ord_state, x = 'State',y= 'Orders')
plt.show()
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
plt.show()
```

### Marital Status

```python
mar_status = sns.countplot(data = df, x = 'Marital_Status')

for m in mar_status.containers:
    mar_status.bar_label(m)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum()
print(sales_state)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
plt.show()
```

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for b in ax.containers:
    ax.bar_label(b)
```

```python
sales_occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
print(sales_occ)
sns.set(rc={'figure.figsize':(20,5)})
occ=sns.barplot(data = sales_occ, x = 'Occupation',y= 'Amount')
plt.show()

```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
pro = sns.countplot(data = df, x = 'Product_Category')

for bars in pro.containers:
    pro.bar_label(bars)
```

```python
sales_pro = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
print(sales_pro)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_pro, x = 'Product_Category',y= 'Amount')
plt.show()
```

```python
sales_ord = df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
print(sales_ord)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_ord, x = 'Product_Category',y= 'Orders')
plt.show()
```

```python
sales_ord = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
print(sales_ord)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_ord, x = 'Product_ID',y= 'Orders')
plt.show()
```

```python
sales_ord = df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
print(sales_ord)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_ord, x = 'Product_ID',y= 'Amount')
plt.show()
```
