```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
netflix=pd.read_csv("netflix_titles.csv")
netflix.head(1)
```

```python
print(netflix.columns)
```

```python
# import sqlchemy and create a sqlite engine
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)

# export the dataframe as a table 'playstore' to the sqlite engine
netflix.to_sql("netflix", con=engine)
```

```python
sql="select * from netflix limit 5"

df_sql=pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Total Number of Movies and TV Shows: Count the total number of movies and TV shows in the dataset?

sql="select type,  count(*) as total_count from netflix group by type ; "


df_sql=pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Number of Titles per Category: Display the count of titles in each category (e.g., Comedy, Drama, Action, etc.).

sql = """SELECT title, COUNT(*) AS title_count
    FROM netflix
    GROUP BY title;"""

df_sql=pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# How many movies and TV shows were released each year?

sql= """select release_year, type, count(*) as title_count
        from netflix
        group by release_year, type
        order by release_year, type limit 10 ; """ 

df_sql=pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# What are the top 10 countries with the most content available on Netflix?

sql= """SELECT country, COUNT(*) AS title_count
    FROM netflix
    GROUP BY country
    ORDER BY title_count DESC
    LIMIT 10;"""

df_sql=pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# What is the distribution of content ratings (e.g., PG, PG-13, R) on Netflix?

sql=""" SELECT rating, COUNT(*) AS rating_count
    FROM netflix
    GROUP BY rating
    order by rating_count desc
    limit 10;"""
    
df_sql=pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Which directors have the most content available on Netflix?

sql = """SELECT director, COUNT(*) AS title_count
    FROM netflix
    WHERE director IS NOT NULL
    GROUP BY director
    ORDER BY title_count DESC
    limit 10;"""

df_sql=pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# What is the average duration of movies and TV shows on Netflix?

sql= """SELECT type, AVG(duration) AS average_duration
    FROM netflix
    GROUP BY type;"""

df_sql=pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# What are the oldest and newest movies/TV shows available on Netflix?

sql= """
    SELECT title, type, release_year
    FROM netflix
    WHERE release_year = (SELECT MIN(release_year) FROM netflix); """

old_movie= pd.read_sql_query(sql, con=engine)
old_movie
```

```python
sql= """
    SELECT title, type, release_year
    FROM netflix
    WHERE release_year = (SELECT MAX(release_year) FROM netflix ); """

new_movie= pd.read_sql_query(sql, con=engine)
new_movie
```

```python
# Longest Movie/TV Show: Find the longest movie and the longest TV show in terms of duration.

longest_movie= """ SELECT title, type, duration
    FROM netflix
    WHERE type = 'Movie'
    ORDER BY duration DESC
    LIMIT 5;"""
lon_mov= pd.read_sql_query(longest_movie, con=engine)
lon_mov
```

```python
smallest_movie= """
    SELECT title, type, duration
    FROM netflix
    WHERE type = 'TV Show'
    ORDER BY duration DESC
    LIMIT 5;"""
sma_mov= pd.read_sql_query(smallest_movie, con=engine)
sma_mov
```

```python
# Titles Added Each Year: Display the number of titles added to Netflix each year.

title_add = """
    SELECT release_year, COUNT(*) AS title_count
    FROM netflix
    GROUP BY release_year
    ORDER BY release_year;"""

tit_add= pd.read_sql_query(title_add, con=engine)
tit_add
```

```python
# 15.	What is the distribution of release date top 10 for Netflix content?
release_date= """
    SELECT strftime(date_added ) AS release_month, COUNT(*) AS title_count
    FROM netflix
   
    GROUP BY date_added
    ORDER BY date_added desc
    limit 10;
"""

release_date= pd.read_sql_query(release_date, con=engine)
release_date
```

```python
 length = """SELECT AVG(LENGTH(description)) AS average_description_length
FROM netflix;"""
    
length = pd.read_sql_query(length, con=engine)
length
```

```python
# top 10 directors with listed_in on netflix file.

top_dic=  """select director, listed_in
            from netflix
            group by director
            order by director desc
            limit  10"""

df= pd.read_sql_query(top_dic, con=engine)
df
```

```python
# Titles with Most Seasons: List the TV shows with the most seasons.

sql=  """
    SELECT title, seasons
    FROM netflix
    WHERE type = 'TV Show'
    ORDER BY seasons DESC
    LIMIT 10;
"""

df_sql=pd.read_sql_query(sql, con=engine)
df_sql
```
